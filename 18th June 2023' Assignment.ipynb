```python
1. What is the role of the 'else' block in a try-except statement? Provide an example scenario where it would be useful.
- Else block in exception handling is executed when there is no exception raised within the try block. 
- The purpose of this block is to define the code inside the try block is executed successfully without any error.
Example:
    
def addition():
    try:
        x = int(input("Enter x : "))
        y = int(input("Enter y : "))
        z = x + y
        print("Result : ", z)
    except ValueError:
        print("Invalid input. Please enter integers only.")
    else:
        print("Code inside try block is executed successfully")
addition()       
```

```python
2. Can a try-except block be nested inside another try-except block? Explain with an example.
- Yes, a try-except block can be nested inside another try-except block this is called nested exception handling.
- It allows for more fine-grained handling of exceptions and the ability to handle exceptions at different levels of code.
Example:
    
def nested_exception() :   
    try:# Outer try block
        num1 = int(input("Enter the numerator: "))  
        try:  # Inner try block
            num2 = int(input("Enter the denominator: "))
            result = num1 / num2
            print("Result:", result)
        except ValueError:
            print("Value Error invalid denominator. Please enter a number.")
        except ZeroDivisionError:
            print("Cannot divide by zero.")
        finally:
            print("Inner try block executed.")
    except ValueError:
        print("Value Error invalid denominator. Please enter a number.")
    finally:
        print("Outer try block executed.")
nested_exception()
```

```python
3. How can you create a custom exception class in Python? Provide an example that demonstrates its usage.
A custom exception class can be created by defining a new class that inherits from the built-in Exception class or any of its subclasses.
Example:

class NotPalindromeException(Exception):
    pass

def check_palindrome(string):
    if string.lower() != string.lower()[::-1]:
        raise NotPalindromeException(f"'{string}'is not a palindrome")

try:
    input_string = input("Enter a string: ")
    check_palindrome(input_string)
    print(f"'{input_string}' is a palindrome.")
except NotPalindromeException as e:
    print("Error:", str(e))
except Exception as e:
    print("An unexpected error occurred:", str(e))
```

```python
7. What are log formatters in Python logging, and how can you customise the log message format using formatters?
-Log formatters are used to define the structure and content of log messages. 
-This format, which shows the level, name, and message separated by a colon (:).
-The default output format that can be configured to include things like timestamp, line number, and other details.

Example

import logging

# Configure the logging settings
logging.basicConfig(level=logging.INFO)

# Create a custom log formatter
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')

# Create a file handler and set the formatter
file_handler = logging.FileHandler('error1.log')
file_handler.setFormatter(formatter)


logger = logging.getLogger()
logger.addHandler(file_handler)


logger.info('This is an informational message')
```
